```python
# 3. 양수, 음수, 0 판별하기 (if~elif~else)

#정수를 입력받아 양수, 음수 또는 0을 출력하는 프로그램을 작성하시오.

num = int(input('숫자를 입력해주세요'))

if num > 0:
        print("양수")
elif num < 0:
        print("음수")
else: print(0)

 
```

```python
# 4. 20대 출입 가능 여부
# 두 사람의 나이를 입력받는다.
# 두 사람 모두 20대(20~29세)이면 출입 가능을 출력하고,
# 그 외에는 출입 불가를 출력하는 프로그램을 작성하시오.

num1 = int(input('첫 번째 나이를 입력해주세요'))
num2 = int(input('두 번째 나이를 입력해주세요'))

if (20 <= num1 <= 29) and (20 <= num2 <= 29):
    print('20대 입니다. 출입 가능') 
else: print('출입 불가!!!')    



```

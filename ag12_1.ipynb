```python
numbers = [15, 8, 23, 7, 31, 18]
target = 23

count = 0

for i in range(len(numbers)):
    count += 1
    if numbers[i] == target:
        print("찾은 위치 :", i)
        break # 찾으면 남은 원소는 비교하지않고 break

print("비교 횟수 :", count)
```

```python
numbers = [15, 8, 23, 7, 31, 18]

max_num = numbers[0]
count = 0

for n in numbers:
    count += 1
    if n > max_num:
        max_num = n

print("최댓값 :", max_num)
print("비교 횟수 :", count)
```

```python
numbers = [5, 3, 8, 1, 4]

# 데이터 5개
# 1회전 : 4번 비교
# 2회전 : 3번 비교
# 3회전 : 2번 비교
# 4회전 : 1번 비교

count = 0

for j in range(len(numbers) - 1):
    for i in range(len(numbers) - 1 - j):
        count += 1
        if numbers[i] > numbers[i + 1]:
            numbers[i], numbers[i + 1] = numbers[i + 1], numbers[i] # 우측 패킹 -> 대입 -> 좌측 언패킹
            # temp = numbers[i]
            # numbers[i] = numbers[i+1]
            # numbers[i+1] = temp

print(numbers)
print("비교 횟수 :", count)
```
